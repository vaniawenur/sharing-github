```python
import cv2
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('D:/vanias/sem 4/PCD/mner jimmy/Tugas 1/burano.jpg')
```

```python
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap='gray')
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(img_hsv, cmap='hsv')
```

```python
gambar = cv2.imread('D:/vanias/sem 4/PCD/mner jimmy/Tugas 1/tcinta.jpg')
```

```python
plt.imshow(gambar)
```

```python
cv2.rectangle(gambar, pt1 = (800, 470), pt2 = (980, 530),color = (255, 0, 0), thickness = 5) 
plt.imshow(gambar)
```

```python
cv2.circle(gambar, center = (950, 50), radius = 50, color = (0, 0, 255), thickness = 5) 
plt.imshow(gambar)
```

```python
cv2.putText(gambar, text = "the Wall of Love", org = (250, 250), fontFace = cv2.FONT_HERSHEY_DUPLEX, fontScale = 2, color = (0, 255, 0), thickness = 2, lineType = cv2.LINE_AA) 
plt.imshow(gambar)
```
